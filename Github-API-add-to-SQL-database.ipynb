```python
#Function to create SQL table and database
def create_sql_table(SQL_TABLE_GITHUB, SQL_TABLE_GITHUB_DEF):
    try:
        #connect to server
        conn = pymysql.connect(host='localhost', user = 'root', password = 'smr@22394') #connection establishment
        cursor = conn.cursor() #cursor with connection

        query = "CREATE DATABASE IF NOT EXISTS " + SQL_DB ##This is SQL query to create a database
        print(query) #printing it to verify
        cursor.execute(query) ##.execute method is to execute the query that we wrote above, cursor does that
        
        query = "CREATE TABLE IF NOT EXISTS " + SQL_DB + "." + SQL_TABLE_GITHUB + " " + SQL_TABLE_GITHUB_DEF + ";"
        print(query)  #printing it to verify
        cursor.execute(query) ##.execute method is to execute the query that we wrote above, cursor does that
        cursor.close() #Close the cursor and connection after we have created the database and table
        conn.close()
        return

    except IOError as e:
        print(e) #exception handling, print the error
        cursor.close() #Close the cursor and connection even if there was error
        conn.close()
```

```python
##Name of db and table
SQL_DB = "Github2"
SQL_TABLE_GITHUB = "github2"
#define the table, this is definition as per MYSQL syntax
SQL_TABLE_GITHUB_DEF = (
"(login VARCHAR(50), id INT NOT NULL PRIMARY KEY, location VARCHAR(250), "
"email VARCHAR(50), hireable VARCHAR(50), bio VARCHAR(1000), twitter_username VARCHAR(50), public_repos INT, "
"public_gists INT, followers INT, following INT, created_at DATETIME)"
)
create_sql_table(SQL_TABLE_GITHUB, SQL_TABLE_GITHUB_DEF) ##calling the above funtion to create the table
```

```python
#Now that the table is created, we need to add rows to it which are to be fetched from API as we did in part 3
```

```python
#connect to server
conn = pymysql.connect(host='localhost', user = 'root', password = 'smr@22394', database = SQL_DB) #connection establishment with the particular db
cursor = conn.cursor()  
#fetch html and convert to json
try:
    for item in list_json:
        urls = item["url"]
        page_each_user = requests.get(urls,headers = headers)
        user_json = page_each_user.json()
        #below line is written, because bio contains "'" single quotes in the string which gave error while fetching them in table 
        #it is replacing one single quote to two single quotes so that SQL understands its a part of the string
        bio = user_json["bio"].replace("'", "''") if user_json["bio"] else user_json["bio"] 
        ##Insert statement as per SQL syntax
        statement = f"""
        INSERT INTO {SQL_TABLE_GITHUB}

        (login, id, location, email, hireable, bio, twitter_username, public_repos,
        public_gists, followers, following, created_at)

        VALUES 
        (
        '{user_json["login"]}','{user_json["id"]}', '{user_json["location"]}', '{user_json["email"]}',
        '{user_json["hireable"]}', '{bio}', '{user_json["twitter_username"]}','{user_json["public_repos"]}',
        '{user_json["public_gists"]}','{user_json["followers"]}','{user_json["following"]}', STR_TO_DATE('{user_json["created_at"]}', "%Y-%m-%dT%H:%i:%sZ")
        );                        
        """
        #Note in the 'created_at' column, I used str_to_date function to convert the string to datetime
        # print(statement)
        cursor.execute(statement)
        time.sleep(1) #wait after each fetch
    conn.commit() #commit is like saving. It commits to the inserts executed above
    cursor.close() #Close the cursor and connection after we have created the database and table
    conn.close()
except IOError as e:
    print(item)
    print(e)
    cursor.close() #Close the cursor and connection even if there was error
    conn.close()
```

5. Optimize your code in (4) to allow for quick look-ups of "login", "location", and "hireable".  I.e., I would like, for example, to run the command  <<  SELECT * FROM table WHERE location = "Tokyo"  >>  fast.  What choices do you make and why? <br>
ANSWER: I would choose to index the three columns "login", "location", and "hireable". By indexing, the search time will be reduced and the query will run faster as it will search using binary search. Hence the complexity reduces from n to log n

```python
#connect to server
conn = pymysql.connect(host='localhost', user = 'root', password = 'smr@22394', database = SQL_DB) #connection establishment with the particular db
cursor = conn.cursor() 
try:
    # Primary index is automatically set up in SQL on the primary key
    # We are creating secondary indexes on login, location and hireable
    # SQL Statement to create a secondary index   
    sqlCreateSecondaryIndex1 = f"""CREATE INDEX index_login ON {SQL_TABLE_GITHUB} (login);"""
    sqlCreateSecondaryIndex2 = f"""CREATE INDEX index_location ON {SQL_TABLE_GITHUB} (location);"""
    sqlCreateSecondaryIndex3 = f"""CREATE INDEX index_hireable ON {SQL_TABLE_GITHUB} (hireable);"""

    #Execute the sql - create secondary index
    cursor.execute(sqlCreateSecondaryIndex1)
    cursor.execute(sqlCreateSecondaryIndex2)
    cursor.execute(sqlCreateSecondaryIndex3)

    #Now list the indexes on the table
    sqlShowIndexes = f"""show index from {SQL_TABLE_GITHUB}""";
    cursor.execute(sqlShowIndexes)

    indexList = cursor.fetchall()
    print(indexList) # Printing the list of indexes on the table
    
    cursor.close() #Close the cursor and connection even if there was error
    conn.close()
    
except Exception as e:
    print(e)
    cursor.close() #Close the cursor and connection even if there was error
    conn.close()
```
